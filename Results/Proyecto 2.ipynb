En la figura 1, se observa el diagrama eléctrico del circuito desarrollado para aplicar a la señal una ganancia programable, la cual posee 6 escalas ajustables, ya sea para aumentar la señal o atenuarla. El circuito cuenta con 4 etapas, la primera es para amplificar la señal de entrada, la segunda es un offset para trabajar en la región lineal del optoacoplador, la tercera etapa es el OTA junto con el Optoacoplador para aislar la señal y por último, la etapa de ganancia progamada mencionada anteriarmente.



             Figura 1. Diagrama eléctrico de la ganancia, acondicionamiento y procesamiento de la señal. 

En la primera etapa del circuito se tiene un amplificador diferencial el cual cuenta en su entrada con un filtro pasa-bajos RC el cual posee una frecuencia de corte teorica de 6.7Hz con el objetivo de eliminar el ruido producido por la termocupla, ya que esta al no contar con aislante recibe mucha interferencia que desestabiliza las proporciones de la realimentacion positiva y negativa del circuito, haciendo que el circuito empiece a oscilar. A su vez este diseño cuenta con una ganancia de 488 para amplificar el voltaje que proviene de la entrada diferencial de la termocupla, esto no establece la relacion entre voltaje y temperatura en 20 mV por grado centigrado. En la figura 2 se puede ver dicho diseño. 



                                Figura 2. Etapa Diferencial.

            Figura 3. Etapa de Offset.

Para la ganancia programable se ha utilizado un divisor de tensión en la entrada y un amplificador operacional con la topología de un no inversor para amplificar la señal o atenuarla según se elija la resistencia de entrada, en la figura 5, se puede notar más de cerca la configuración de resistencias de este apartado. 



            Figura 5. Etapa de ganancia.


Las ganancias y atenuaciones se pueden analizar con la siguiente ecuación:

\begin{equation}
Ganancia = \frac{1k\Omega}{(R\Omega+1k\Omega)}(1+\frac{10k\Omega}{(1k\Omega)})
\end{equation}
Donde R puede tomar el valor de R4, R5, R6, R7, R8, R9.  

Con R4; \begin{equation}
Ganancia = \frac{1k\Omega}{(1M\Omega+1k\Omega)}(1+\frac{10k\Omega}{1k\Omega}) = 0.01
\end{equation}

Con R5; \begin{equation}
Ganancia = \frac{1k\Omega}{(33k\Omega+1k\Omega)}(1+\frac{10k\Omega}{1k\Omega}) = 0.32
\end{equation}

Con R6; \begin{equation}
Ganancia = \frac{1k\Omega}{(15k\Omega+1k\Omega)}(1+\frac{10k\Omega}{1k\Omega}) = 0.69
\end{equation}

Con R7; \begin{equation}
Ganancia = \frac{1k\Omega}{(10k\Omega+1k\Omega)}(1+\frac{10k\Omega}{1k\Omega}) = 1
\end{equation}

Con R8; \begin{equation}
Ganancia = \frac{1k\Omega}{(4.7k\Omega+1k\Omega)}(1+\frac{10k\Omega}{1k\Omega}) = 1.93
\end{equation}


Con R9; \begin{equation}
Ganancia = \frac{1k\Omega}{(10\Omega+1k\Omega)}(1+\frac{10k\Omega}{1k\Omega}) = 10.89
\end{equation}

Con lo cual se cumplen las 6 escalas de la señal, además si se toma como base una señal de 1 V de prueba, esta ira de los 10 mV a los 10 V.

Para mostrar la señal en tiempo real, se ha utilizado una pantalla LCD conectado al Arduino y de este modo confirmar el funcionamiento de la programación.


Haciendo un barrido de voltaje en la simulación se observa que a partir de de 4.5 V el voltaje en la entrada del OTA con respecto a la salida del Optoacoplador es lineal. En la figura 6 se puede observar dicho comportamiento.



                  Figura 6. Barrido DC en la entrada del OTA con respecto a la salida del Optoacoplador.

A partir de esta grafica podemos observar que la region que presenta un compportamiento lineal entre la entrada del OTA y la salida del optoacoplador es de 4.5V en adelante, llegando hasta los 10V de entrada en el OTA, valor que corresponde a los voltajes de alimentacion para los amplificador operacionales. Exportando los datos y trabajandolo un poco en excel, obtuvimos que la variacion entre un punto y otro es de 4mV, no obstante, observando solo los valores que representan un aumento de 1 grado centigrado, podemos ver que la variacion es de 7.5mV aproximadamente.
